```python
def eval_comp(user_answer,sign,numb):
        if sign == '<=':
            return user_answer<=numb;
        elif sign == '<':
            return user_answer <numb;
        elif sign == '=':
            return user_answer ==numb
        elif sign == '>':
            return user_answer >numb;
        elif sign == '>=':
            return user_answer>=numb;
```
